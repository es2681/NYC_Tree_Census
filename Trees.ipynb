```python
# Import dependencies 
import pandas as pd
```

## Group trees by species and borough

```python
# Load the 2015 Street Trees Census CSV file
file_to_load = "2015StreetTreesCensus_TREES.csv"
trees_df = pd.read_csv(file_to_load)
trees_df
```

```python
# Include only relevant columns in DataFrame
trees_df = trees_df[['tree_id', 'status', 'health', 'spc_latin', 'spc_common', 'zipcode', 'boroname', 'nta_name']]
trees_df
```

```python
# Sort data for alive trees only
trees_df = trees_df.loc[trees_df['status'] == 'Alive']
trees_df = trees_df.drop(columns=['status'])
trees_df
```

```python
# Create a DataFrame for trees in the borough of Manhattan
manhattan = trees_df.loc[trees_df['boroname'] == 'Manhattan']
manhattan
```

```python
# Create a DataFrame for trees in the borough of Brooklyn
brooklyn = trees_df.loc[trees_df['boroname'] == 'Brooklyn']
brooklyn
```

```python
# Create a DataFrame for trees in the borough of Queens
queens = trees_df.loc[trees_df['boroname'] == 'Queens']
queens
```

```python
# Create a DataFrame for trees in the borough of The Bronx
bronx = trees_df.loc[trees_df['boroname'] == 'Bronx']
bronx
```

```python
# Create a DataFrame for trees in the borough of Staten Island
staten_island = trees_df.loc[trees_df['boroname'] == 'Staten Island']
staten_island
```

```python
# Group the trees for each borough by species and count the number of each species per borough
mn_trees = manhattan.groupby(['spc_common'])['spc_common'].count().sort_values(ascending=False)
bk_trees = brooklyn.groupby(['spc_common'])['spc_common'].count().sort_values(ascending=False)
qn_trees = queens.groupby(['spc_common'])['spc_common'].count().sort_values(ascending=False)
bx_trees = bronx.groupby(['spc_common'])['spc_common'].count().sort_values(ascending=False)
si_trees = staten_island.groupby(['spc_common'])['spc_common'].count().sort_values(ascending=False)
```

```python
# Create a trees list
trees_list = trees_df[['spc_latin', 'spc_common']].drop_duplicates()
# Remove trees that don't have any name 
trees_list = trees_list.dropna()
trees_list
```

```python
# Transform the grouped tree data for each borough into individual DataFrames
si_trees = pd.DataFrame(si_trees)
si_trees.columns = ['SI Count']
si_trees = si_trees.reset_index()

mn_trees = pd.DataFrame(mn_trees)
mn_trees.columns = ['NY Count']
mn_trees = mn_trees.reset_index()

bk_trees = pd.DataFrame(bk_trees)
bk_trees.columns = ['BK Count']
bk_trees = bk_trees.reset_index()

bx_trees = pd.DataFrame(bx_trees)
bx_trees.columns = ['BX Count']
bx_trees = bx_trees.reset_index()

qn_trees = pd.DataFrame(qn_trees)
qn_trees.columns = ['QN Count']
qn_trees = qn_trees.reset_index()
```

```python
# Merge borough tree count DataFrames into a single DataFrame for all of NYC
df = pd.merge(trees_list, mn_trees, how='left', on=['spc_common', 'spc_common'])
df = pd.merge(df, bk_trees, how='left', on=['spc_common', 'spc_common'])
df = pd.merge(df, qn_trees, how='left', on=['spc_common', 'spc_common'])
df = pd.merge(df, bx_trees, how='left', on=['spc_common', 'spc_common'])
df = pd.merge(df, si_trees, how='left', on=['spc_common', 'spc_common'])
df
```

```python
# Change NaN's to 0
df = df.fillna(0)
df
```

```python
# Verify data types
df.dtypes
```

```python
# Change floats to integers
df['NY Count'] = df['NY Count'].astype('int64')
df['SI Count'] = df['SI Count'].astype('int64')
df.dtypes
```

```python
# confirm DataFrame
df.head(15)
```

```python
# Save DataFrame to CSV
df.to_csv("NYC_Trees_Count.csv", index=False)
```

## Determine the number of trees per 1,000 people per zipcode

```python
# Load population by zipcode CSV file
file_to_load = "population_by_zip.csv"
population_df = pd.read_csv(file_to_load)
population_df
```

```python
# Remove zipcodes where population is 0 (i.e. zipcodes for Central Park, World Trade Center, etc.)
population_df = population_df[population_df['pop'] != 0]
population_df
```

```python
# Group the number of trees by zipcode
zip_trees = trees_df.groupby(['zipcode'])['zipcode'].count()
zip_trees = pd.DataFrame(zip_trees)
zip_trees.columns = ["Zip Count"]
zip_trees = zip_trees.reset_index()
zip_trees
```

```python
# inner merge the population and tree data 
ny = pd.merge(zip_trees, population_df, how='inner', left_on=['zipcode'], right_on=['zip'])
ny
```

```python
# Drop the redundant zip column
ny = ny.drop(columns=['zip'])
ny
```

```python
# Calculate the number of trees per person
trees_per_person = ny['Zip Count'] / ny['pop']
trees_per_person
```

```python
# Add a column to the DataFrame for number of trees per 1,000 people
ny['Trees per 1000 People'] = ny['Zip Count'] / ny['pop'] * 1000
ny['Trees per 1000 People'] = ny['Trees per 1000 People'].map("{:.2f}".format)
ny
```

```python
# Sort table by zipcodes with the greatest number of trees per person
ny = ny.sort_values(by='Trees per 1000 People', ascending=False)
ny
```
